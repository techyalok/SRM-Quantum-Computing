```python
import qiskit
print(qiskit.__version__)
```

```python
from qiskit import QuantumCircuit, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector as sv, Operator as op
from qiskit.visualization import plot_bloch_multivector as pbm, plot_histogram as ph, plot_state_qsphere as psq
from numpy import sqrt # mainly used for 1/sqrt(n) in superposition of n states in a qubit
from qiskit.visualization import array_to_latex

air = AerSimulator()
```

# Exp - 3 - Multi Qubit Gates

```python
qc = QuantumCircuit(3) 
for qubit in range(3): 
    qc.h(qubit)
qc.draw()
```

```python
final_state = sv.from_instruction(qc)
array_to_latex(final_state, prefix="\\text{Statevector} = ")
```

```python
pbm(final_state)
```

```python
qc.measure_all()
counts = air.run(qc).result().get_counts()
ph(counts)
```

```python
qc = QuantumCircuit(2) 
qc.h(0)
qc.x(1)
qc.draw()
```

```python
stve = sv.from_instruction(qc)
pbm(stve)
```

```python
qc.measure_all()
counts = air.run(qc).result().get_counts()
ph(counts)
```

```python
qc = QuantumCircuit(2) 
qc.h(0)  
qc.cx(0,1) 
qc.draw()
```

```python
matrix = op.from_circuit(qc).to_matrix()
array_to_latex(matrix)
```

```python
steve = sv.from_instruction(qc)
pbm(steve)
```

```python
psq(steve)
```

```python
qc.measure_all()
counts = air.run(qc).result().get_counts()
ph(counts)
```
